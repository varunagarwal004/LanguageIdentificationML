```python
import pandas as pd
from sklearn.neural_network import MLPClassifier
from sklearn.svm import SVC

from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
from sklearn.feature_extraction import DictVectorizer

from sklearn.pipeline import Pipeline
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.model_selection import GridSearchCV, ParameterGrid

import numpy as np

import warnings
warnings.filterwarnings("ignore")

from google.colab import files
import io
data = files.upload()
```

```python
gender_submission = pd.read_csv(io.StringIO(data['gender_submission.csv'].decode('utf-8')))
df_test = pd.read_csv(io.StringIO(data['test.csv'].decode('utf-8')))
df_train = pd.read_csv(io.StringIO(data['train.csv'].decode('utf-8')))
```

```python
#creating model with the training set
from sklearn.neural_network import MLPClassifier
#building classifier based on sex, age
y = df_train.loc[:, 'Survived']
sex_pclass = df_train.loc[:, ['Sex', 'Pclass']]
sex_pclass.replace(['male', 'female'], [0, 1], inplace=True)
x = sex_pclass

classifier = MLPClassifier(hidden_layer_sizes=(10,), activation='logistic', solver='lbfgs', random_state=42).fit(x, y)
pred_y = classifier.predict(x)

accuracy = accuracy_score(y, pred_y)
# accuracy
ids = df_train.loc[:, 'PassengerId'].values
sex_pclass_prediction_vals = {'PassengerId': ids, 'Survived': pred_y}
sex_pclass_prediction = pd.DataFrame(data=sex_pclass_prediction_vals).set_index('PassengerId')
```

```python
print(accuracy)
```

```python
#making test set pred
# y = df_test.loc[:, 'Survived']
sex_pclass_test = df_test.loc[:, ['Sex', 'Pclass']]
ids_test = df_test.loc[:, 'PassengerId'].values
sex_pclass_test.replace(['male', 'female'], [0, 1], inplace=True)
x_test = sex_pclass_test

pred_y_test = classifier.predict(x_test)
pred_y_test

sex_pclass_prediction_test_vals = {'PassengerId': ids_test, 'Survived': pred_y_test}
sex_pclass_prediction_test = pd.DataFrame(data=sex_pclass_prediction_test_vals).set_index('PassengerId')
sex_pclass_prediction_test.to_csv('prediction.csv')
```
